```python
def handle_multi_level_columns(df):
    # Create a list to hold the new column names
    new_columns = []

    # Iterate over the current column names
    for col in df.columns:
        # If level 0 is unnamed or 'Unnamed: 0_level_0', leave level 1 as is
        if str(col[0]).lower().startswith('unnamed'):
            new_columns.append(col[1].replace(' ', '_').upper())
        else:  # Otherwise, write the column name as {level0}_{level1}
            new_columns.append(f'{col[0]}_{col[1]}'.replace(' ', '_').upper())

    # Assign the new column names to the DataFrame
    df.columns = new_columns

    return df

player_stats=pd.read_html('https://www.hockey-reference.com/leagues/NHL_2023_skaters.html')
new_df=pd.DataFrame(player_stats[0])
handle_multi_level_columns(new_df)
```

```python
def ensure_columns(df, cols):
    for col in cols:
        if col not in df.columns:
            df[col] = np.nan
    return df[cols]

def handle_multi_level_columns(df):
    # Create a list to hold the new column names
    new_columns = []

    # Iterate over the current column names
    for col in df.columns:
        # If level 0 is unnamed or 'Unnamed: 0_level_0', leave level 1 as is
        if str(col[0]).lower().startswith('unnamed'):
            new_columns.append(col[1].replace(' ', '_').upper())
        else:  # Otherwise, write the column name as {level0}_{level1}
            new_columns.append(f'{col[0]}_{col[1]}'.replace(' ', '_').upper())

    # Assign the new column names to the DataFrame
    df.columns = new_columns

    return df

start_year = 2023
print(start_year)

url_skaters=f'https://www.hockey-reference.com/leagues/NHL_{start_year}_skaters.html'
url_goalies=f'https://www.hockey-reference.com/leagues/NHL_{start_year}_goalies.html'
df_skaters = pd.read_html(url_skaters)
df_goalies = pd.read_html(url_goalies)
df_skaters_cleaned = pd.DataFrame(df_skaters[0])
df_goalies_cleaned = pd.DataFrame(df_goalies[0])
df_skaters_cleaned = handle_multi_level_columns(df_skaters_cleaned)
df_goalies_cleaned = handle_multi_level_columns(df_goalies_cleaned)
df_skaters_cleaned = df_skaters_cleaned.drop_duplicates(['RK'], keep='first')
df_goalies_cleaned = df_goalies_cleaned.drop_duplicates(['RK'], keep='first')
df_skaters_cleaned['SEASON'] = start_year
df_goalies_cleaned['SEASON'] = start_year
df_skaters_cleaned.to_csv("skaters_1996_2023.csv", index=False)
df_goalies_cleaned.to_csv("goalies_1996_2023.csv", index=False)

skaters_columns = df_skaters_cleaned.columns.to_list()
goalies_columns = df_goalies_cleaned.columns.to_list()

start_year -= 1
years = list(range(1996, start_year + 1))
years.reverse()
years.remove(2005) # lockout season, no data
time.sleep(10)

for year in years:
    print(year)
    url_skaters=f'https://www.hockey-reference.com/leagues/NHL_{start_year}_skaters.html'
    url_goalies=f'https://www.hockey-reference.com/leagues/NHL_{start_year}_goalies.html'
    df_skaters = pd.read_html(url_skaters)
    df_goalies = pd.read_html(url_goalies)
    df_skaters_cleaned = pd.DataFrame(df_skaters[0])
    df_goalies_cleaned = pd.DataFrame(df_goalies[0])
    df_skaters_cleaned = handle_multi_level_columns(df_skaters_cleaned)
    df_goalies_cleaned = handle_multi_level_columns(df_goalies_cleaned)
    df_skaters_cleaned = df_skaters_cleaned.drop_duplicates(['RK'], keep='first')
    df_goalies_cleaned = df_goalies_cleaned.drop_duplicates(['RK'], keep='first')
    df_skaters_cleaned['SEASON'] = year
    df_goalies_cleaned['SEASON'] = year

    df_skaters_cleaned = ensure_columns(df_skaters_cleaned, skaters_columns)
    df_goalies_cleaned = ensure_columns(df_goalies_cleaned, goalies_columns)

    df_skaters_cleaned.to_csv("skaters_1996_2023.csv", index=False, header=False, mode='a')
    df_goalies_cleaned.to_csv("goalies_1996_2023.csv", index=False, header=False, mode='a')
    time.sleep(10)
```
